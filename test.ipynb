```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data for ATS Score Breakdown
categories = [
    "Role Match", "Experience & Achievements", "Skills Match", "Education Fit",
    "Readability & Grammar", "Formatting & ATS Compliance", "Keyword Density & Buzzword Balance", "Action Verbs & Impactful Language"
]
scores = [80, 85, 75, 65, 90, 95, 70, 80]

# Setting up the plot style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
colors = sns.color_palette("viridis", len(scores))

# Creating bar chart
ax = sns.barplot(y=categories, x=scores, palette=colors, orient='h')
plt.xlabel("Score (%)", fontsize=12, fontweight='bold')
plt.ylabel("Categories", fontsize=12, fontweight='bold')
plt.title("ATS Score Breakdown", fontsize=14, fontweight='bold')

# Displaying values on bars
for i, (score, category) in enumerate(zip(scores, categories)):
    ax.text(score + 1, i, f"{score}%", ha='left', va='center', fontsize=10, fontweight='bold')

plt.xlim(0, 100)
plt.tight_layout()
plt.show()

# Displaying Final Recommendations
recommendations = """
Final Recommendations:
1. **Major Improvement Area**: The resume lacks specific details about the Data Scientist role and its requirements.
   - Improve by inserting relevant keywords and phrases from the job description into the summary, skills, and achievements sections.

2. **Education Section**:
   - The education section is too lengthy and includes unnecessary details.
   - Limit it to only the most relevant information and remove percentages and minor details.

3. **Additional Tip**:
   - Consider adding a summary or objective statement at the beginning of the resume to provide a brief overview of your experience and skills as a Data Scientist.

**Final ATS Score**: 84
"""

print(recommendations)
```
